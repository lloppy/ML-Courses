```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize = (9, 5))
```

```python
x = np.arange(-10, 11)
y = x**2

plt.plot(x, y)
```

```python
plt.grid()
```

```python
# Enable gridlines
plt.grid()

# Plot the data
plt.plot(x, y)

# Display the plot
plt.show()
```

```python
plt.plot(x, y)
plt.grid()
plt.title('Grafic $y=x^2$') #latex

plt.xlabel('x')
plt.ylabel('y')
```
