5. Дан список чисел: 1 5 2 4 3. Выведите все элементы списка, которые больше предыдущего элемента.

````python
numbers = [1, 5, 2, 4, 3]
ans = []

for i in range(0, len(numbers) - 1):
  if numbers[i] < numbers[i + 1]:
    ans.append(numbers[i + 1])
ans
````

6. Дано действительное положительное число a и целоe число n.

Вычислите a^n. Решение оформите в виде функции power(a, n).

Стандартной функцией возведения в степень пользоваться нельзя.

````python
def power(a, n):
  ans = a
  for i in range(1, n):
    ans *= a
  return ans

print(power(2, 50))
````

7. Написать функцию season, принимающую 1 аргумент — номер месяца (от 1 до 12), и возвращающую время года, которому этот месяц принадлежит (зима, весна, лето или осень).

````python
def season(month):
  if month in {12, 1, 2}:
    return("Зима")
  elif month in {3, 4, 5}:
    return("Весна")
  elif month in {6, 7, 8}:
    return("Лето")
  elif month in {9, 10, 11}:
    return("Осень")
  else:
    return("Wrong input")

print(season(12))
````

10. Написать функцию arithmetic, принимающую 3 аргумента: первые 2 - числа, третий - операция, которая должна быть произведена над ними. Если третий аргумент +, сложить их; если —, то вычесть; * — умножить; / — разделить (первое на второе). В остальных случаях вернуть строку "Неизвестная операция".
*) Измените функцию, чтобы при вызове arithmetic(x,y) выполнялась операция сложения

````python
def arithmetic(a, b, operator = "+"):
  if operator == "+": return a + b
  elif operator == "-": return a - b
  elif operator == "*": return a * b
  elif operator == ":": return a / b
  else:
    return("Неизвестная операция")

print(arithmetic(5, 7))
print(arithmetic(5, 7, "*"))
````
