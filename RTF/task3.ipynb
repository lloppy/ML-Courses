# Задание 3


```python
a = '1, 2, 3, 4'
b = '10, 20, 30, 40'

v1 = np.array([float(x.strip()) for x in a.split(',')])
v2 = np.array([float(x.strip()) for x in b.split(',')])

v3 = v1 + v2
print(v3)

# V4 Numpy вектор с покоординатными произведениями каждого второго числа V1 на каждое второе число V2, развёрнутого в обратном порядке
v4 = v1[1::2] * ( v2[1::2][::-1] )
print(v4)
```
