# pd.Series

Одномерный набор данных. Похож на `np.ndarray`, но с более расширенным функционалом

### Создание Series из np.array

```python
import numpy as np
np_array = np.random.randint(low=-15, high=20, size=4) # генерация рандомных значений
np_array
```

### Преобразование np.array к Series

```python
pd_series_1 = pd.Series(np_array) # преобразование Series из np.array
pd_series_1
```

### Создание Series из списков и словарей

```python
# создаем серию значений из списка
s = pd.Series([10, 11, 12, 13, 14])
s
```

```python
# создаем объект Series из словаря
s = pd.Series({'Mike': 'Dad',
               'Marcia': 'Mom',
               'Mikael': 'Son',
               'Bleu': 'Best' })
s
```

### Создание Series с помощью скалярного значения

```python
# создаем объект Series, состоящий из одного элемента
s = pd.Series(2)
s
```

### Название индексов

```python
# Так же можно аннотировать индексы, чтобы потом было более удобно обращаться к элементам
pd_series_2 = pd.Series(np_array, index=['1st day', '2nd day', '3rd day', '4th day'])
pd_series_2
```

### Получение значения по индексу

```python
pd_series_2['4th day']
```

### Название Series

```python
# А еще можно дать pd.Series имя, чтобы было совсем красиво
pd_series_3 = pd.Series(np_array, index=['1st day', '2nd day', '3rd day', '4th day'], name='Temperature')
pd_series_3
```

### Операции как с list
с `series` можно работать как с `list`

```python
pd_series_3[0]
```

```python
pd_series_3[1:3]
```

```python
pd_series_3[::-1]
```

### Сортировка
`series` можно отсортировать как по значениям `sort_values()`, так и по индексу  `sort_index()`

```python
pd_series_3.sort_index()
```

```python
pd_series_3.sort_values()
```

### Операции как с np.ndarray

с `series` можно работать как с `np.ndarray`

```python
pd_series_3 + 100
```

```python
np.exp(pd_series_3)
```

```python
term_1 = pd.Series(np.random.randint(0, 10, 5))
term_2 = pd.Series(np.random.randint(0, 10, 6))
```

```python
print('Term1:\n{}'.format(term_1))
```

```python
print('Term2:\n{}'.format(term_2))
```

```python
print('Sum:\n{}'.format(term_1 + term_2))
```

### shape

```python
term_1.shape
```

### date_range

```python
date_range = pd.date_range(start='2020/01/01', periods=10)
pd_series_4 = pd.Series(np.random.rand(10) * 25, index=date_range)
pd_series_4
```

### Особенности операций

```python
pd_series_4 > 10
```

```python
(pd_series_4 > 10).sum()
```

```python
pd_series_4[pd_series_4 > 10] # в качестве индекса можно указать выражение, и нам будут возвращены только те элементы, для которых значение является
```

```python
pd_series_4[(pd_series_4 > 10) & (pd_series_4 < 15)] # предопределенные операторы для побитовых операций с числами
```

# DataFrame

Двумерная таблица данных. Имеет индекс и набор столбцов (возможно, имеющих разные типы).

```python
# Dataframe можно составить из словаря. Ключ будет соответсовать названию колонки
d={'one':pd.Series([1,2,3],index=['a','b','c']),
   'two':pd.Series([1,2,3,4],index=['a','b','c','d'])}
df=pd.DataFrame(d)
df
```

### Вывод названия индексов

```python
df.index
```

### Вывод названия колонок

```python
df.columns
```

### Обращение к значениям

```python
df['one'] # ['c']
```

```python
df.one
```

```python
df['one']['c']
```

```python
df.one.c
```

Если указать диапазон индексов, то это означает диапазон строк. Причём **последняя строка включается в таблицу.**

Диапазон целых чисел даёт диапазон строк с такими номерами, **не включая последнюю строку** (как обычно при индексировании списков).

```python
df['b':'d']
```

```python
df[1:3]
```

### Операции

```python
df['three'] = df['one'] * df['two']
df
```

### concat

```python
df2=pd.DataFrame(
    {'one':{'e':0,'f':1},
     'one_tr':{'f':2}
    })
df2
```

```python
pd.concat([df,df2], sort=False)
```

```python
df
```

### Метод astype
позволяет изменить тип колонки

```python
df['two'] = df['two'].astype('float64')
```

```python
df
```

### Сортировка по значению какого-нибудь из признаков

```python
df.sort_values(by='one', ascending=False)
```

### Сортировка по значению группы признаков

```python
df.sort_values(by=['one', 'two'], ascending=[True, False])
```

### Метод map

```python
d = { 1 : 100}
df['one'] = df['one'].map(d)
df
```

### Метод replace

```python
df = df.replace({'two': d})
df
```

### Поиск значений NaN в объектах библиотеки pandas

```python
# какие элементы являются значениями NaN?
df.isnull()
```

```python
# подсчитываем количество значений NaN в каждом столбце
df.isnull().sum()
```
